```python
import pandas as pd
import numpy as np
ננ
df = pd.DataFrame({
    'CustomerID': np.arange(1, 11),
    'FirstName': ['John', 'Steve', 'Anna', 'Mike', 'Emily', 'Jake', 'Laura', 'Nick', 'Olivia', 'David'],
    'LastName': ['Smith', 'Johnson', 'Williams', 'Jones', 'Brown', 'Davis', 'Miller', 'Wilson', 'Moore', 'Taylor'],
    'Age': np.random.randint(25, 65, size=10),
    'TotalPurchaseAmount': np.random.randint(100, 1000, size=10)
})

print("Original DataFrame:")
print(df)

df['classification'] = df['TotalPurchaseAmount'].apply(lambda x: 'Low' if x < 500 else 'High')
print("\nDataFrame with 'classification' column:")
print(df)

def classify_by_age(row):
    if row['classification'] == 'Low' and row['Age'] < 20:
        return 'Low_Young'
    elif row['classification'] == 'High' and row['Age'] >= 20:
        return 'HIGH_Old'
    elif row['classification'] == 'Low' and row['Age'] >= 20:
        return 'Low_Old'
    elif row['classification'] == 'High' and row['Age'] < 20:
        return 'High_Young'

df['classification_by_age'] = df.apply(classify_by_age, axis=1)
print("\nDataFrame with 'classification_by_age' column:")
print(df)
```

```python
df_sorted_by_name = df.sort_values(by=['FirstName'])
print("\nDataFrame sorted by FirstName (A-Z):")
print(df_sorted_by_name)

df_sorted = df.sort_values(by=['TotalPurchaseAmount', 'Age'], ascending=[True, False])
print("\nDataFrame sorted by TotalPurchaseAmount (ascending) and Age (descending):")
print(df_sorted)
```

```python
np.random.seed(100)
df = pd.DataFrame(np.random.randint(0, 100, size=(20, 10)), columns=list('ABCDEFGHIJ'))
n_rows, n_cols = df.shape
row_idx = np.random.randint(0, n_rows)
col_idxs = np.random.choice(n_cols, size=n_cols//2, replace=False)
df.iloc[row_idx, col_idxs] = np.nan
n_nan_remaining = 15 - len(col_idxs)
row_idxs = np.random.randint(0, n_rows, size=n_nan_remaining)
col_idxs = np.random.randint(0, n_cols, size=n_nan_remaining)
df.iloc[row_idxs, col_idxs] = np.nan

print("Original DataFrame with NaN values:")
print(df)

df_copy = df.copy()

df_no_nan = df_copy.dropna()
print("\nDataFrame after dropping rows with NaN:")
print(df_no_nan)

df_filled_random = df_copy.fillna(np.random.randint(0, 100))
print("\nDataFrame after filling NaN with random values:")
print(df_filled_random)

df_copy['C'] = df_copy['C'].fillna(df_copy['C'].mean())
df_copy['D'] = df_copy['D'].fillna(df_copy['D'].mean())
print("\nDataFrame after filling NaN in 'C' and 'D' with mean values:")
print(df_copy)

df_filtered = df_copy.dropna(thresh=len(df_copy.columns) - 1, axis=0, subset=df_copy.columns.drop('D'))
print("\nDataFrame after dropping rows with at least 2 NaN values (excluding 'D'):")
print(df_filtered)
```
